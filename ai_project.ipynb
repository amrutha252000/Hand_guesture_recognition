```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import datasets, layers, models, losses, Model
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
     'leapGestRecog',
     validation_split=0.05,
     subset="training",
     seed=123,
     image_size=(160, 160),
     batch_size=32,
    
)

val_ds = tf.keras.utils.image_dataset_from_directory(
  'leapGestRecog',
  validation_split=0.05,
  subset="validation",
  seed=123,
  image_size=(160, 160),
  batch_size=32,
)

normalization_layer = tf.keras.layers.Rescaling(1./255)

def change_inputs(images, labels):
    x = tf.image.resize(normalization_layer(images),[160,160], method=tf.image.ResizeMethod.NEAREST_NEIGHBOR)
    
    new_labels = tf.one_hot(labels, depth=10)
    return x, new_labels

train_ds = train_ds.map(change_inputs)
val_ds = val_ds.map(change_inputs)
```
